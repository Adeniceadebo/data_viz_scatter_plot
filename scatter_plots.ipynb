```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_theme()
```

```python
df_income = pd.read_csv('income_and_age.csv')
df_income
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df_income, x='Income', y='Age', hue='Sex', palette=['#E02828', '#272635'])
plt.title('Relationship Between Income and Age Among Survey Respondents', fontweight='bold', fontsize=14)
plt.xlabel('Income (in USD)', fontsize=12, fontweight='bold')
plt.ylabel('Age (in years)', fontsize=12, fontweight='bold')
plt.legend(title='Legend', fontsize=10, loc ='upper right')
plt.grid(True)
plt.savefig('income_vs_age.png', dpi=300, bbox_inches='tight')
plt.show()

```
